*   CSE_ID is a special search engine ID configured for image search.

```python
import requests
import os
import time
import hashlib
from google.colab import drive

drive.mount('/content/drive')


API_KEY = 'AIzaSyArqJwCByEdt5tBwYfV8BqSXz9HuKjWU14'
CSE_ID = '536840746d9c74e82'


folder_name = '/content/drive/MyDrive/images'


def google_search(search_term, api_key, cse_id, num=10, start=1):
    url = f"https://www.googleapis.com/customsearch/v1?q={search_term}&key={api_key}&cx={cse_id}&searchType=image&num={num}&start={start}"
    response = requests.get(url).json()
    return response


def save_image(url, folder, idx):
    if not os.path.exists(folder):
        os.makedirs(folder)


    filename = hashlib.md5(url.encode('utf-8')).hexdigest()
    file_path = f"{folder}/{filename}.jpg"


    if os.path.exists(file_path):
        print(f"Gambar {filename} sudah ada, lewati penyimpanan.")
        return

    try:
        img_data = requests.get(url).content
        with open(file_path, 'wb') as img_file:
            img_file.write(img_data)
        print(f"Gambar {idx + 1} disimpan di {file_path}")
    except Exception as e:
        print(f"Error saat menyimpan gambar {idx + 1}: {e}")


search_term = "tahu kupat"
total_images_to_fetch = 80
images_per_request = 10

for start_index in range(1, total_images_to_fetch + 1, images_per_request):
    results = google_search(search_term, API_KEY, CSE_ID, num=images_per_request, start=start_index)

    if 'items' in results:
        for idx, item in enumerate(results['items']):
            img_url = item['link']
            save_image(img_url, folder_name, start_index + idx - 1)
    else:
        print("Tidak ada gambar yang ditemukan atau pencarian selesai.")
        break


    time.sleep(1)
```
