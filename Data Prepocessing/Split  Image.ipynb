```python
import shutil
import os

def remove_folders(folders_to_remove, base_directory):
    """
    Menghapus beberapa folder di dalam direktori utama.
    
    Parameters:
        folders_to_remove (list): List nama folder yang akan dihapus.
        base_directory (str): Path ke direktori utama.
    """
    for folder_name in folders_to_remove:
        folder_path = os.path.join(base_directory, folder_name)
        
        # Cek jika folder ada
        if os.path.exists(folder_path) and os.path.isdir(folder_path):
            shutil.rmtree(folder_path)
            print(f"Folder {folder_path} has been removed.")
        else:
            print(f"Folder {folder_path} does not exist.")

# Daftar nama folder yang ingin dihapus
folders_to_remove = ['baklava', 'panna_cotta', 'carrot_cake','peking_duck','seaweed_salad', 'onion_rings','edamame','creme_brulee', 'nachos','miso_soup']  # Ganti dengan folder yang ingin dihapus

# Path ke direktori utama
dataset_path = "gambar"  # Ganti dengan path direktori Anda

# Panggil fungsi untuk menghapus folder
remove_folders(folders_to_remove, dataset_path)

print("Folder removal is complete.")
```

```python
dataset_path = "gambar"
output_path = "ML_NutriCheck"


train_ratio = 0.8
validation_ratio = 0.1
test_ratio = 0.1


all_class_name = [d for d in os.listdir(dataset_path) if os.path.isdir(os.path.join(dataset_path, d))]

train_data = {}
validation_data = {}
test_data = {}


for class_name in all_class_name:
    category_path = os.path.join(dataset_path, class_name)
    files = os.listdir(category_path)
    random.shuffle(files)

    total_files = len(files)
    train_split = int(train_ratio * total_files)
    validation_split = int(validation_ratio * total_files)

    train_data[class_name] = files[:train_split]
    validation_data[class_name] = files[train_split:train_split + validation_split]
    test_data[class_name] = files[train_split + validation_split:]


def move_files(files, source_path, destination_path):
    for file in files:
        source_file = os.path.join(source_path, file)
        destination_file = os.path.join(destination_path, file)
        if os.path.isdir(source_file):
            shutil.copytree(source_file, destination_file)
        else:
            shutil.copyfile(source_file, destination_file)


for category, files in train_data.items():
    source_path = os.path.join(dataset_path, category)
    destination_path = os.path.join(output_path, "train", category)
    os.makedirs(destination_path, exist_ok=True)
    move_files(files, source_path, destination_path)


for category, files in validation_data.items():
    source_path = os.path.join(dataset_path, category)
    destination_path = os.path.join(output_path, "validation", category)
    os.makedirs(destination_path, exist_ok=True)
    move_files(files, source_path, destination_path)


for category, files in test_data.items():
    source_path = os.path.join(dataset_path, category)
    destination_path = os.path.join(output_path, "test", category)
    os.makedirs(destination_path, exist_ok=True)
    move_files(files, source_path, destination_path)

print("Dataset division is complete. The dataset has been divided into train, validation, and test folders.")
```

```python
def count_files_in_class_name(path):
    categories = os.listdir(path)
    counts = {}
    for category in categories:
        category_path = os.path.join(path, category)
        if os.path.isdir(category_path):
            counts[category] = len(os.listdir(category_path))
    return counts

output_path = "ML_NutriCheck"

train_counts = count_files_in_class_name(os.path.join(output_path, 'train'))
validation_counts = count_files_in_class_name(os.path.join(output_path, 'validation'))
test_counts = count_files_in_class_name(os.path.join(output_path, 'test'))

print("Amount of data in train folder:", train_counts)
print()
print("Amount of data in validation folder:", validation_counts)
print()
print("Amount of data in test folder:", test_counts)
```

```python
import os
import shutil

# Batas jumlah file
train_limit = 800
validation_limit = 100
test_limit = 100

# Folder dataset
train_folder = "ML_NutriCheck/train"
validation_folder = "ML_NutriCheck/validation"
test_folder = "ML_NutriCheck/test"

# Fungsi untuk menghapus file jika jumlahnya melebihi batas
def remove_excess_files(folder, limit):
    for class_name in os.listdir(folder):
        class_path = os.path.join(folder, class_name)
        if os.path.isdir(class_path):
            files = os.listdir(class_path)
            if len(files) > limit:
                excess_files = files[limit:]  # File yang lebih dari batas
                for file in excess_files:
                    file_path = os.path.join(class_path, file)
                    os.remove(file_path)
                    print(f"File {file_path} telah dihapus.")

# Hapus file di folder train, validation, dan test
remove_excess_files(train_folder, train_limit)
remove_excess_files(validation_folder, validation_limit)
remove_excess_files(test_folder, test_limit)

print("Penghapusan file selesai.")
```
